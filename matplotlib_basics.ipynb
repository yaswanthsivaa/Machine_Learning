```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

**Simple Line Plot**

```python
x = np.linspace(0,10,50)
y = np.sin(x)

plt.plot(x,y)
plt.show()
print(y)
```

```python
plt.figure(figsize=(7,4))
plt.plot(x,y,color='red',marker='x',linestyle="--")
plt.title("customized line plot")
plt.xlabel("x-values")
plt.ylabel("y-values")
plt.grid(True)
plt.show()
```

```python
students = ['Alice', 'Bob', 'Charlie', 'David']
marks = [85,92,78,88]

plt.bar(students, marks, color='red')
plt.title("- Students Marks -")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.show()
```

```python
plt.barh(students, marks, color='orange')
plt.xlabel('Marks')
plt.ylabel('Students')
plt.show()
```

```python
# Scatter Plot

np.random.seed(10)
x = np.random.randn(5)
y = np.random.randn(5)

plt.scatter(x, y, color='purple')
plt.title('Scatter Plot')
print(x)
print(y)
plt.show()
```

```python
data = np.random.randn(1000)
bins = [-3, -2, -1, 0, 1, 2, 3]

plt.hist(data, bins=bins)
plt.show()
```

```python
plt.hist(data, bins=15, color='skyblue')
plt.title("Data Distribution (Histogram)")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
# Pie Chart
plt.figure(figsize=(5,5))
labels = ['Python', 'Java', 'CPP', 'R']
sizes = [45, 35, 15, 5]
explodes = [0.2, 0, 0, 0]
colors = ['pink', 'black', 'lightgreen', 'skyblue']
plt.pie(sizes, labels=labels, explode=explodes, autopct="%1.1f%%",colors=colors, shadow=True)
# explode is to separate  one or more wedges(slices) from the center of the pie chart
# sizes must be first parameter
# colors we can give our own colors  -- if don't give it takes its default colors
plt.title("Proportion:")
plt.show()
```

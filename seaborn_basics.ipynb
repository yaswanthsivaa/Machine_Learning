```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
print(sns.__version__)
```

**Seaborn Provides some built-in Datasets**

```python
tips = sns.load_dataset('tips')
```

```python
tips.shape
```

```python
tips.head()
```

```python
# set global styles & color palette
sns.set_style("darkgrid")
sns.set_palette("viridis")
```

**Histogram - Distribution of Total bill**

```python
plt.figure(figsize=(6,4))
sns.histplot(tips['total_bill'], kde=True, color='teal') # kde = kernal density exploration, used to draw the curve
plt.title("Distribution of Total bill")
plt.xlabel("Total Bill")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x='day', y='total_bill', data=tips, palette='pastel', hue='day', legend=False)
plt.title("Total Bill By day")
plt.show()
```

```python
# box plot is used to understand the presence of any outliers in the dataset

plt.figure(figsize=(6,4))
sns.countplot(x='day', data=tips, hue='sex', palette='Set2')
plt.title("Total Bill Day")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.scatterplot(x='total_bill', y='tip', data=tips, hue='sex',size='size', palette='cool')
plt.title('Bill Vs Tip Relationship')
plt.show()
```

**Pair-Plot**

```python
sns.pairplot(tips, hue='sex', diag_kind='kde')
plt.suptitle("Pair-wise relationship")
plt.show()
```

**HeatMap - Feature Correlation**

```python
correlation_matrix = tips.corr(numeric_only=True)
plt.figure(figsize=(6,4))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidth=0.5)
plt.title("Feature Correlation Heatmap")
plt.show()
```
